```python
#Food ordering system

#Admin Commands
import json
class Restaurant:
    def __init__(self):
        self.foods={}
        self.food_id=len(self.foods)
    def food_items(self):
        self.name=input("Enter food Name:")
        self.quantity=int(input("Enter food Quantity: "))
        self.price=int(input("Enter food Price :$ "))
        self.discount=int(input("Enter food Discount :$ "))
        self.stock=int(input("Enter food Stock :Kg "))
        self.item={"name":self.name,"quantity":self.quantity,"price":self.price,"discount":self.discount,"stock":self.stock}
        self.food_id=len(self.foods)+1
        self.foods[self.food_id]=self.item
        print(self.foods)
        with open("Food_Items.Json","w") as f:
            json.dump(self.foods,f)
        print("---Item has been added!---")
    def remove_food_items(self):                   #Deleting current food items
        del self.foods[int(input("Enter the food_id which you would like to delete:"))]
        print("Remaining food items are:",self.foods)
        with open("Food_Items.Json","w") as f:
            json.dump(self.foods,f)
            print("Remaining food items are:",self.foods)
            
    def edit_food_items(self):                    #Edit current food items
        f_id=int(input("Enter the food_id which you would like to edit:"))
        for i in self.foods[f_id]:
            self.foods[f_id][i]=input(f"Enter the {i} you'd like to update the item with:")
        print(self.foods)
        with open("Food_Items.Json","w") as f:
            json.dump(self.foods,f)
        print("Food_Item updated successfully")
        
    def view_food_items(self):                  #View list of food items
        print("List of the current food items")
        for i in self.foods:
            print("food_id :", i,"\n")
            for j in self.foods[i]:
                print(j,":",self.foods[i][j])
            
x=Restaurant()
x.food_items()
x.food_items()
x.view_food_items()
```
